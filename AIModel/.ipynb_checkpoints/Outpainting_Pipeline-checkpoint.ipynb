```python
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from diffusers import StableDiffusionInpaintPipeline

```

```python
pipe = StableDiffusionInpaintPipeline.from_pretrained(
    "runwayml/stable-diffusion-inpainting",
    torch_dtype=torch.float16
).to("cuda")

pipe.safety_checker = None
```

```python
image = Image.open("images/img2.jpg").convert("RGB")

plt.figure(figsize=(6,6))
plt.imshow(image)
plt.axis("off")

width, height = image.size
print(width, height)
```

```python
new_width = width + 256
new_height = height

# create black canvas
new_image = Image.new("RGB", (new_width, new_height), (0,0,0))

# paste original image
new_image.paste(image, (0,0))

plt.figure(figsize=(6,6))
plt.imshow(new_image)
plt.axis("off")
```

```python
mask = np.zeros((new_height, new_width), dtype=np.uint8)

# region to generate
mask[:, width:] = 255

mask_image = Image.fromarray(mask)

# resize first
new_image = new_image.resize((1024,1024))
mask_image = mask_image.resize((1024,1024), Image.NEAREST)

plt.figure(figsize=(6,6))
plt.imshow(mask_image, cmap="gray")
plt.axis("off")

print(new_image.size)
print(mask_image.size)
```

```python
prompt = "photorealistic indoor background continuation"

result = pipe(
    prompt=prompt,
    image=new_image,
    mask_image=mask_image,
    num_inference_steps=50,
    guidance_scale=7.5
).images[0]
```

```python
plt.figure(figsize=(6,6))
plt.imshow(result)
plt.axis("off")
```
